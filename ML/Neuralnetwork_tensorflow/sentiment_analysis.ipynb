```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.imdb.load_data(num_words=10000)
print(len(X_train))
print(len(X_train[0]),y_train[0])
```

```python
X_train=keras.preprocessing.sequence.pad_sequences(X_train,maxlen=256,value=0,padding='pre',truncating='pre')
X_test=keras.preprocessing.sequence.pad_sequences(X_test,maxlen=256,value=0,padding='pre',truncating='pre')
print(X_train.shape)
print(X_test.shape)
```

```python
model=keras.Sequential([keras.layers.Embedding(10000,16,input_length=256),
                        keras.layers.GlobalAveragePooling1D(),  
                        keras.layers.Dense(16,activation='relu'),
                        keras.layers.Dense(1,activation='sigmoid')])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=10,batch_size=512,validation_split=0.2,verbose=1)
loss,accuracy=model.evaluate(X_test,y_test)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```

```python
train_loss, train_acc = model.evaluate(X_train, y_train)
print(f"Train Accuracy: {train_acc*100:.2f}%")
print(f"Test Accuracy: {accuracy*100:.2f}%")
print(f"Difference: {(train_acc-accuracy)*100:.2f}%")
```
