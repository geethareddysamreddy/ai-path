```python
import pandas as pd
import numpy as np
```

```python
d=pd.DataFrame({"Age":[20,25,30,35,40]})
print(d.describe())
```

```python
def Outlier(nums):
    for i in nums:
        if i>100:
            print(f"Outlier value:{i}")
Outlier([10,20,30,40,1000])
```

```python
##Clipping
df=pd.DataFrame({"Marks":[10,20,30,40,1000]})
df["Marks"]=df["Marks"].clip(upper=100)
print(df)
```

```python
##Normalization (Linear Scaling)
N=((d["Age"]-d["Age"].min())/(d["Age"].max()-d["Age"].min()))
print(N)
```

```python
##Z-Score Normalization
Z=(d["Age"]-d["Age"].mean())/d["Age"].std()
print(Z)
```

```python
##Log Scaling
l=np.log(d["Age"])
print(l)
```

```python
##Binning
d["Age_Binned"] = pd.cut(d["Age"], bins=3,labels=["Teen", "Adult", "Old"])

print(d["Age_Binned"])  
```

```python
##quantile Binning
d=pd.DataFrame({"age":[10,20,30,40,50,60,70,80,90]})
d["Age_Quantile_Binned"] = pd.qcut(d["age"], q=3,labels=["Teen", "Adult", "Old"])
print(d["Age_Quantile_Binned"])
```

```python
#Data cleaning
n=pd.DataFrame({"Name":["John","Alice",np.nan,"Bob","Charlie","Geetha"],
                "Age":[25,30,np.nan,35,40,np.nan]})
print(n.isnull().sum())
n["Age"].fillna(n["Age"].mean(),inplace=True)
print(n)
print(n.dropna())
```

```python
#duplicate removal
s=pd.DataFrame({"Name":["John","Alice","Bob","Bob","Charlie","Geetha"]})
s=s.drop_duplicates()
print(s)
```

```python
##Polynomial Transform
p=pd.DataFrame({"x":[1,2,3,4,5]})
p["x_squared"]=p["x"]**2
print(p)
```
