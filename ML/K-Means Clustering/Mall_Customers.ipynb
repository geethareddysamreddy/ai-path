```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = {
    'Income':[15,16,17,18,19,70,72,74,75,76],
    'Spending':[39,81,6,77,40,90,88,95,85,89]
}
df=pd.DataFrame(data)
print(df.head())
```

```python
kmeans=KMeans(
    n_clusters=2,
    random_state=42
)
kmeans.fit(df)
```

```python
clusters=kmeans.labels_
print(clusters)
```

```python
df['Clusters']=clusters
print(df)
print(df.columns)
df.drop('Clusters', axis=1)
```

```python
plt.scatter(
    df['Income'],
    df['Spending'],
    
)
plt.xlabel('Income')
plt.ylabel('Spending')
plt.title('Customer Segment')
plt.show()
```

```python
print(kmeans.cluster_centers_)
```

```python
#Elbow method
X=df[['Income', 'Spending']]
wcss = []

for i in range(1, 11):

    model = KMeans(
        n_clusters=i,
        random_state=42
    )

    model.fit(X)

    wcss.append(model.inertia_)

print(wcss)
```

```python
##Plot Elbow Curve
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('No.of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
##Silhouette Score
from sklearn.metrics import silhouette_score

score = silhouette_score(df, clusters)

print(score)
```
