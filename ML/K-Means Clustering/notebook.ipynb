```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\bharath simha reddy\Desktop\genai_journey\ML\K-Means Clustering\data\Mall_Customers.csv")
print(df)

```

```python
X=df.iloc[:,[3,4]].values
wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,random_state=42,init='k-means++')
    model.fit(X)
    wcss.append(model.inertia_)
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('no.of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Curve')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df=pd.read_csv(r"C:\Users\bharath simha reddy\Desktop\genai_journey\ML\K-Means Clustering\data\Mall_Customers.csv")
print(df)
X=df.iloc[:,[3,4]].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

centroids = scaler.inverse_transform(kmeans.cluster_centers_)

plt.figure(figsize=(8, 6))
for cluster in range(5):
    plt.scatter(
        X[y_kmeans == cluster, 0],
        X[y_kmeans == cluster, 1],
        s=50,
        alpha=0.6,
        label=f'Cluster {cluster + 1}'
    )

plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='black', marker='X', label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering (k=5)')
plt.legend()
plt.show()
```
